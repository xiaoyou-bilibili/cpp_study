## 18.1 复习前面介绍的c++11功能



```c++
// c++提供了多种声明
// 比如我们可以使用auto自动推断类型
auto a = 111;
cout<<a;
```

```c++
// 可以使用decltype来把变量类型声明为表达式指定类型
double x = 11.2;
// 这里声明一个变量y，类型和x保持一致
decltype(x) y;
y = 2.4;
cout<<y;
```

```c++
// 函数可以返回指定类型
double f1(double,int);
// 下面这个和上面是等价的
auto f2(double, int) -> double;
```

```c++
// 模板别名using
// 以前我们使用typedef来进行定义
typedef std::vector<std::string>::iterator itType;
// 下面这个就是新用法了。这个新用法和旧方法的差别就是新语法可以用于模板部分的具体化
using itType = std::vector<std::string>::iterator;
```

```c++
// 空指针nullptr
// 以前是使用0来表示空指针，现在使用nullptr来表示空指针
```

```c++
// c++新增了unique_ptr,shared_ptr,weak_ptr这三种指针类型
```

```c++
// c++提供了一个不抛出异常的关键词
void f3() noexcept;
```

```c++
// 传统的c++枚举类比较低级，c++11新增了一种枚举，可以使用class或者struct来定义
enum old1 {yes,no,maybe};
enum class new1{never, sometimes, often, always};
enum struct new2{never,lever,server};
```

```c++
// c++提供了关键词 explicit来禁止单参数构造函数导致的自动转换
```

```c++
//c++ 可以在类定义中直接初始化成员
class Session{
    int men=10;
    double men2{19966.54};
}
```

```c++
// c++提供了基于范围的for循环
double prices[5] = {4.99,10.99,6.87,7.99,8.49};
for(double x:prices)
    cout<<x<<endl;
```

```c++
// 可以直接使用auto快速推断
for(auto x:prices)
    cout<<x<<endl;
```

```c++
// 右值引用
// 传统的c++引用成为左值引用
int n;
int *pt = new int;
const int b = 101; // 这里我们定义好b后就不能修改了
const int &rb = b; // 但是我们可以直接去修改
```

```c++
// c++新增了右值引用 右值引用可关联到右值，即可出现在赋值表达式右边，但不能对其应用地址运算符的值。
int x = 10;
int y=23;
int && r2 =x+y;
```

## 18.3 新的类功能

```c++
// c++新增了复制构造函数和移动构造函数
someclass::someclass(const someclass &);// 复制构造函数
someclass::someclass(someclass &&);// 移动构造函数
```

```c++
// 我们可以使用default来显示的声明某些方法的默认版本
class a{
    public:
    a()=default;
    a(a &&)=delete; // 关键词delete可以禁用某一种方法
}
```
